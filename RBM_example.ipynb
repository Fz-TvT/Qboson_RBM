```python
from RBM import BernoulliRBM_all
" 可选择的简单model LogisticRegression LDA SVM GaussianNB(朴素贝叶斯) MultinomialNB（多项式贝叶斯） MLPClassifier(多层感知机) Perctron(线性感知机) DecisionTreeClassifier决策树"
modelname="LogisticRegression"  # 选择model
rbm = BernoulliRBM_all(n_components=100, learning_rate=0.1, random_state=0, verbose=True, model_name=modelname)
X_train, X_test, y_train, y_test=rbm.load_data()
```

```python
model = rbm.train_model(X_train, y_train, modelname) # 仅仅训练model
# 在测试集上评估模型
print("\nTest Set Performance:")
rbm.evaluate_model(model, X_test, y_test)
# 可视化部分预测结果
rbm.visualize_predictions(model, X_test, y_test)
```
