```python
from Sample_verify import Sample
"T=3.4时和noise=0.19对应"
noise=0.19
T=3.4
a=0.8
b=0.1
NN=90000 ##ising 和 mcmc采样的采样次数
sam=Sample(noise,T,NN)
N=10 #定义二维spin方格的形状，N*N为自旋变量个数
sam.generate_data(N) #生成数据并保存进入matrix.txt文件 
```

```python
import numpy as np
import matplotlib.pyplot as plt
array=np.loadtxt(r'matrix.txt',delimiter=' ', dtype=np.float64) #edge list
data2=sam.mcmc_sample(NN,array) ##采样mcmc采样
numbers,P=sam.calculate(data2)
data=sam.ising_sample(array,a,b) ##采样ising采样
numbers1,P1=sam.calculate(data)
plt.plot(numbers1, P1, 'r-', label='Cim sample')
plt.scatter(numbers, P,c='blue',s=10,label='Mcmc sample')
plt.xlabel('Energy', fontsize=14)
plt.ylabel('Probabilities', fontsize=14)
# 显示图例
plt.legend()
```
